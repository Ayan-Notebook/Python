```python
import matplotlib.pyplot as plt

language = ['Python', 'Java', 'C++', 'JavaScript', 'C#']
sizes = [40, 25, 15, 10, 10]
colors = ['Black', 'Green', 'pink', 'yellow', 'grey']


plt.pie(sizes, labels=language, colors=colors, autopct='%1.1f%%')
plt.title('Programming language use')

plt.show()
```

```python
import matplotlib.pyplot as plt


ages = [25, 30, 35, 40, 45, 50, 55, 60]
salaries = [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000]


plt.scatter(ages, salaries, color='blue',  edgecolors='black')


plt.title('Salary vs Age')
plt.xlabel('Age')
plt.ylabel('Salary')


plt.grid(True)
plt.show()
```

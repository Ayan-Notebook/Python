```python
import matplotlib.pyplot as plt

# creating array for overs and run rate
overs = [10, 20, 30, 40, 50]

run_rate = [17.5, 10.6, 9, 12.5, 18]

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(overs, run_rate, marker='o', linestyle='-')

# Adding labels and title
plt.title('India\'s Over-wise Run Rate in a 50-overs One Day Match')
plt.xlabel('Overs')
plt.ylabel('Run Rate')

# Display the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data: Sales of different flavors of ice creams
flavors = ['Vanilla', 'Strawberry', 'Chocolate', 'Mint', 'Caramel']
sales = [200, 150, 180, 190, 250]

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(flavors, sales, color=['blue', 'pink', 'brown', 'green', 'orange'])

# Adding labels and title
plt.title('Sales of Different Flavors of Ice Creams')
plt.xlabel('Ice Cream Flavors')
plt.ylabel('Sales (units)')

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Rent', 'Groceries', 'Utilities', 'Entertainment', 'Transportation']
expenses = [7000, 1000, 800, 200, 500]

plt.bar(categories, expenses, color=['red', 'blue', 'green', 'yellow', 'brown'])
plt.xlabel('Categories')
plt.ylabel('Monthly Expenses (in dollars)')
plt.title('Monthly Expenses in Different Spending Categories')
plt.show()
```
